```python
from google.colab.widgets import TabBar
import ipywidgets as widgets
import time
import pandas as pd
from ipywidgets import interact
from ipywidgets import Text
from ipywidgets import DatePicker
from ipywidgets import IntText
from ipywidgets import Label
from datetime import date

tb_all = TabBar(['Introduction', 'Main Menu', 'Retrieve Data', 'Insert Data', 'Update Data', 'Delete Data'])

# a simple function for jumping between tabs
def showPage(tabName):
  with tb_all.output_to(tabName):
    display()

##########################################################
################ Introduction Tab - Begin ################
##########################################################
with tb_all.output_to('Introduction'):

# create 1 label
  intro_label = Label("ISOM5260-Lab07b Sample Code")

# create 1 button
  btnContinue = widgets.Button(description="Continue")

# define the event handlers for the two buttons when they are being clicked
  def on_btnContinue_button_clicked(b):
    showPage("Main Menu")
  btnContinue.on_click(on_btnContinue_button_clicked)

# display the objects
  intro_row1HBox = widgets.HBox([intro_label])
  intro_row2HBox = widgets.HBox([btnContinue])

  display(widgets.VBox([intro_row1HBox, intro_row2HBox]))
################ Introduction Tab - End ################



#######################################################
################ Main Menu Tab - Begin ################
#######################################################
with tb_all.output_to('Main Menu'):

# create 1 label
  mainMenu_label = Label("Please select a function...")

# create 4 button
  btnRetrievePage = widgets.Button(description="Retrieve Data")
  btnInsertPage = widgets.Button(description="Insert Data")
  btnUpdatePage = widgets.Button(description="Update Data")
  btnDeletePage = widgets.Button(description="Delete Data")

# define the event handlers for the two buttons when they are being clicked
  def on_btnRetrievePage_button_clicked(b):
    showPage("Retrieve Data")
  btnRetrievePage.on_click(on_btnRetrievePage_button_clicked)

  def on_btnInsertPage_button_clicked(b):
    showPage("Insert Data")
  btnInsertPage.on_click(on_btnInsertPage_button_clicked)

  def on_btnUpdatePage_button_clicked(b):
    showPage("Update Data")
  btnUpdatePage.on_click(on_btnUpdatePage_button_clicked)

  def on_btnDeletePage_button_clicked(b):
    showPage("Delete Data")
  btnDeletePage.on_click(on_btnDeletePage_button_clicked)

# display the objects
  mainMenu_row1HBox = widgets.HBox([mainMenu_label])
  mainMenu_row2HBox = widgets.HBox([btnRetrievePage])
  mainMenu_row3HBox = widgets.HBox([btnInsertPage])
  mainMenu_row4HBox = widgets.HBox([btnUpdatePage])
  mainMenu_row5HBox = widgets.HBox([btnDeletePage])

  display(widgets.VBox([mainMenu_row1HBox, mainMenu_row2HBox, mainMenu_row3HBox, mainMenu_row4HBox, mainMenu_row5HBox]))
################ Main Menu Tab - End ################



###########################################################
################ Retrieve Data Tab - Begin ################
###########################################################
with tb_all.output_to('Retrieve Data'):

# create 5 textboxes and 1 label
  id_textbox = Text(value='', placeholder='Enter a Student ID', description='Student ID:', disabled=False)
  spacer = Label("---------------------------------------------------------------------------------------------------------------------------------")
  name_textbox = Text(description='Name:', disabled=True)
  email_textbox = Text(description='Email:', disabled=True)
  year_intbox = IntText(description='Year:', disabled=True)
  DOB_datepicker = DatePicker(description='DOB:', disabled=True)

# create 3 buttons
  retrieve_button = widgets.Button(description="Retrieve")
  clear_button = widgets.Button(description="Clear")
  back_button = widgets.Button(description="Back")

# define the event handlers for the two buttons when they are being clicked
  def on_retrieve_button_clicked(b):
    c.execute("SELECT * FROM STUDENT WHERE student_id = '" + id_textbox.value + "'")
    for row in c:
      name_textbox.value = row[1]
      email_textbox.value = row[2]
      year_intbox.value = row[3]
      DOB_datepicker.value = row[4]
  retrieve_button.on_click(on_retrieve_button_clicked)

  def on_clear_button_clicked(b):
    name_textbox.value = ""
    email_textbox.value = ""
    year_intbox.value = -1
    DOB_datepicker.value = None
  clear_button.on_click(on_clear_button_clicked)

  def on_back_button_clicked(b):
    showPage("Main Menu")
  back_button.on_click(on_back_button_clicked)

# display the objects
  line1HBox = widgets.HBox([id_textbox, retrieve_button])
  line2HBox = widgets.HBox([spacer])
  line3HBox = widgets.HBox([name_textbox, email_textbox])
  line4HBox = widgets.HBox([year_intbox, DOB_datepicker])
  line5HBox = widgets.HBox([clear_button])
  line6HBox = widgets.HBox([back_button])

  display(widgets.VBox([line1HBox, line2HBox, line3HBox, line4HBox, line5HBox, line6HBox]))
################ Retrieve Data Tab - End ################



#########################################################
################ Insert Data Tab - Begin ################
#########################################################
with tb_all.output_to('Insert Data'):

# create 5 labels and 5 textboxes
  INS_id_label = Label("Student ID: ")
  INS_id_textbox = Text(value='', placeholder='Student ID', disabled=False)

  INS_name_label = Label("Name: ")
  INS_name_textbox = Text(placeholder='Name:', disabled=False)

  INS_email_label = Label("Email: ")
  INS_email_textbox = Text(placeholder='Email:', disabled=False)

  INS_year_label = Label("Year: ")
  INS_year_intbox = IntText(placeholder='Year:', disabled=False)

  INS_dob_label = Label("DOB")
  INS_DOB_datepicker = DatePicker(placeholder='DOB:', disabled=False)

# create 3 buttons
  INS_add_button = widgets.Button(description="Add Student")
  INS_clear_button = widgets.Button(description="Clear")
  INS_back_button = widgets.Button(description="Back")

# define the event handlers for the two buttons when they are being clicked
  def on_INS_add_button_clicked(b):
    c.execute(
    """INSERT INTO STUDENT VALUES ({}, '{}', '{}', {}, TO_DATE('{}', 'YYYY-MM-DD'))
    """.format(INS_id_textbox.value, INS_name_textbox.value, INS_email_textbox.value, INS_year_intbox.value, INS_DOB_datepicker.value))

    c.execute("commit")
    INS_add_button.disabled = True
    print("Done!")
  INS_add_button.on_click(on_INS_add_button_clicked)

  def on_INS_clear_button_clicked(b):
    INS_id_textbox.value = ""
    INS_name_textbox.value = ""
    INS_email_textbox.value = ""
    INS_year_intbox.value = 0
    INS_DOB_datepicker.value = None
    INS_add_button.disabled = False
  INS_clear_button.on_click(on_INS_clear_button_clicked)

  def on_INS_back_button_clicked(b):
    showPage("Main Menu")
  INS_back_button.on_click(on_INS_back_button_clicked)

# display the objects
  INS_labelVBox = widgets.VBox([INS_id_label, INS_name_label, INS_email_label, INS_year_label, INS_dob_label])
  INS_textboxVBox = widgets.VBox([INS_id_textbox, INS_name_textbox, INS_email_textbox, INS_year_intbox, INS_DOB_datepicker])
  INS_line1HBox = widgets.HBox([INS_labelVBox, INS_textboxVBox])
  INS_line2HBox = widgets.HBox([INS_add_button, INS_clear_button])
  INS_line3HBox = widgets.HBox([INS_back_button])

  display(widgets.VBox([INS_line1HBox, INS_line2HBox, INS_line3HBox]))
################ Insert Data Tab - End ################



#########################################################
################ Update Data Tab - Begin ################
#########################################################
with tb_all.output_to('Update Data'):

# create 6 label and 5 textboxes
  UPD_id_label = Label("Student ID: ")
  UPD_id_textbox = Text(value='', placeholder='Student ID', disabled=False)

  UPD_spacer = Label("---------------------------------------------------------------------------------------------------------------------------------")

  UPD_name_label = Label("Name: ")
  UPD_name_textbox = Text(placeholder='Name:', description='Name:', disabled=False)

  UPD_email_label = Label("Email: ")
  UPD_email_textbox = Text(description='Email:', disabled=False)

  UPD_year_label = Label("Year: ")
  UPD_year_intbox = IntText(description='Year:', disabled=False)

  UPD_dob_label = Label("DOB")
  UPD_DOB_datepicker = DatePicker(description='DOB:', disabled=False)

# create 4 buttons
  UPD_retrieve_button = widgets.Button(description="Retrieve")
  UPD_update_button = widgets.Button(description="Update")
  UPD_clear_button = widgets.Button(description="Clear")
  UPD_back_button = widgets.Button(description="Back")

# define the event handlers for the three buttons when they are being clicked
  def on_UPD_retrieve_button_clicked(b):
    c.execute("SELECT * FROM STUDENT WHERE student_id = '" + UPD_id_textbox.value + "'")
    for row in c:
      UPD_name_textbox.value = row[1]
      UPD_email_textbox.value = row[2]
      UPD_year_intbox.value = row[3]
      UPD_DOB_datepicker.value = row[4]
  UPD_retrieve_button.on_click(on_UPD_retrieve_button_clicked)

  def on_UPD_clear_button_clicked(b):
    UPD_name_textbox.value = ""
    UPD_email_textbox.value = ""
    UPD_year_intbox.value = 0
    UPD_DOB_datepicker.value = None
  UPD_clear_button.on_click(on_UPD_clear_button_clicked)

  def on_UPD_update_button_clicked(b):
    c.execute(
    """
      UPDATE STUDENT
      SET student_name = '{}',
      email = '{}',
      year = {},
      dob = TO_DATE('{}', 'YYYY-MM-DD')
      WHERE student_id = '{}'
    """.format(UPD_name_textbox.value, UPD_email_textbox.value, UPD_year_intbox.value, UPD_DOB_datepicker.value.strftime("%Y-%m-%d"), UPD_id_textbox.value)
    )
    c.execute("commit")
    print("Update Success!")
  UPD_update_button.on_click(on_UPD_update_button_clicked)

  def on_UPD_back_button_clicked(b):
    showPage("Main Menu")
  UPD_back_button.on_click(on_UPD_back_button_clicked)

# display the objects
  UPD_line1HBox = widgets.HBox([UPD_id_label, UPD_id_textbox, UPD_retrieve_button])
  UPD_line2HBox = widgets.HBox([UPD_spacer])
  UPD_textboxVBox = widgets.VBox([UPD_name_textbox, UPD_email_textbox, UPD_year_intbox, UPD_DOB_datepicker])
  UPD_line3HBox = widgets.HBox([UPD_textboxVBox])
  UPD_line4HBox = widgets.HBox([UPD_update_button, UPD_clear_button])
  UPD_line5HBox = widgets.HBox([UPD_back_button])

  display(widgets.VBox([UPD_line1HBox, UPD_line2HBox, UPD_line3HBox, UPD_line4HBox, UPD_line5HBox]))
################ Update Data Tab - End ################



#########################################################
################ Delete Data Tab - Begin ################
#########################################################
with tb_all.output_to('Delete Data'):

# create 6 label and 5 textboxes
  DEL_id_label = Label("Student ID: ")
  DEL_id_textbox = Text(value='', placeholder='Student ID', disabled=False)

  DEL_spacer = Label("---------------------------------------------------------------------------------------------------------------------------------")

  DEL_name_label = Label("Name: ")
  DEL_name_textbox = Text(placeholder='Name:', description='Name:', disabled=False)

  DEL_email_label = Label("Email: ")
  DEL_email_textbox = Text(description='Email:', disabled=False)

  DEL_year_label = Label("Year: ")
  DEL_year_intbox = IntText(description='Year:', disabled=False)

  DEL_dob_label = Label("DOB")
  DEL_DOB_datepicker = DatePicker(description='DOB:', disabled=False)

# create 4 buttons
  DEL_retrieve_button = widgets.Button(description="Retrieve")
  DEL_delete_button = widgets.Button(description="Delete")
  DEL_clear_button = widgets.Button(description="Clear")
  DEL_back_button = widgets.Button(description="Back")

# define the event handlers for the three buttons when they are being clicked
  def on_DEL_retrieve_button_clicked(b):
    c.execute("SELECT * FROM STUDENT WHERE student_id = '" + DEL_id_textbox.value + "'")
    for row in c:
      DEL_name_textbox.value = row[1]
      DEL_email_textbox.value = row[2]
      DEL_year_intbox.value = row[3]
      DEL_DOB_datepicker.value = row[4]
  DEL_retrieve_button.on_click(on_DEL_retrieve_button_clicked)

  def on_DEL_clear_button_clicked(b):
    DEL_name_textbox.value = ""
    DEL_email_textbox.value = ""
    DEL_year_intbox.value = 0
    DEL_DOB_datepicker.value = None
  DEL_clear_button.on_click(on_DEL_clear_button_clicked)

  def on_DEL_delete_button_clicked(b):
    c.execute(
    """
          DELETE FROM STUDENT
          WHERE student_id = '{}'
    """.format(DEL_id_textbox.value)
    )
    c.execute("commit")
    print("Delete Success!")
  DEL_delete_button.on_click(on_DEL_delete_button_clicked)

  def on_DEL_back_button_clicked(b):
    showPage("Main Menu")
  DEL_back_button.on_click(on_DEL_back_button_clicked)

# display the objects
  DEL_line1HBox = widgets.HBox([DEL_id_label, DEL_id_textbox, DEL_retrieve_button])
  DEL_line2HBox = widgets.HBox([DEL_spacer])
  DEL_textboxVBox = widgets.VBox([DEL_name_textbox, DEL_email_textbox, DEL_year_intbox, DEL_DOB_datepicker])
  DEL_line3HBox = widgets.HBox([DEL_textboxVBox])
  DEL_line4HBox = widgets.HBox([DEL_delete_button, DEL_clear_button])
  DEL_line5HBox = widgets.HBox([DEL_back_button])

  display(widgets.VBox([DEL_line1HBox, DEL_line2HBox, DEL_line3HBox, DEL_line4HBox, DEL_line5HBox]))
################ Delete Data Tab - End ################


showPage("Introduction")
```
